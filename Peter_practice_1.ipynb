```python
# Reading the contents of the document into a string

with open('Add19664_whole.txt', 'rt', encoding="utf-8") as myfile2:
    contents=myfile2.read()
print(contents)
```

```python
# Making the above string into a list where sof pasuq is the separator

contents2 = contents.split(':')
print(contents2)
```

```python
# Searching the list for all occurrences of the pronominal suffix הון

for element in contents2:
    q= re.findall(r'.+הון', element)
    if q:
        (print(f"line {contents2.index(element)}", element))
    else:
        (print("No match"))
```

```python
#Splitting a string by multiple patterns:

print(re.split(r'ך|ם|ן|ף|ץ|:', contents))
```

```python
#Searching for words related to the root 'אמר'

for element in contents2:
    print(re.findall('א.?מ.?ר', element))
```
